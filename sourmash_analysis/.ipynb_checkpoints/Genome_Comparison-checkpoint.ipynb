> Download these genomes using same command without the `-l` option for `complete`.

```python
while read i; do mkdir -p complete_genomes_ncbi/$i; ncbi-genome-download bacteria --genus $i -F fasta -o complete_genomes_ncbi/$i; done < incomplete_genomes_list.txt
```

> put all genomes in one folder

```python
for i in complete_genomes_ncbi/*; do mkdir -p $i/assemblies; mv $i/refseq/bacteria/*/GCF_* $i/assemblies/; done
```

```python
for i in complete_genomes_ncbi/*; do rm -r $i/refseq; done # clean up each directory by removing empty sub-directories
```

## Sourmash

```python
sourmash compute -h
```

```python
for i in complete_genomes_ncbi/*; do mkdir -p $i/sourmash; done
```

```python
for i in complete_genomes_ncbi/*/assemblies/GCF_*; do bname=`basename $i`; sourmash compute -k 31 --scaled 10000 -o 
```

```python
while read genus; do for i in complete_genomes_ncbi/$genus/assemblies/GCF_*; do bname=`basename $i`; sourmash compute -k 31 -n 10000 -o complete_genomes_ncbi/$genus/sourmash/${bname}-k31.sig $i; done; done < genome_list.txt 
```

> Ignore the genomes that had `N` characters and failed to generate sketches...for now.

> Make decriptive labels for plotting later

```python
while read genus; do for i in complete_genomes_ncbi/$genus/assemblies/GCF_*; do bname=`basename $i`; acc=`echo $bname | cut -d_ -f2`; echo ${genus}-GCF_${acc} >> complete_genomes_ncbi/$genus/sourmash/genus_accession_labels.txt; done; done < genome_list.txt 
```

> Check the labels

```python
head complete_genomes_ncbi/Acidovorax/sourmash/genus_accession_labels.txt
```

> Loop to comapre signatures in each genus folder

```python
sourmash compare -h
```

```python
for i in complete_genomes_ncbi/*; do bname=`basename $i`; sourmash compare -k 31 -o complete_genomes_ncbi/${bname}/sourmash/${bname}-k31_cmp.csv $i/sourmash/*.sig; done
```
